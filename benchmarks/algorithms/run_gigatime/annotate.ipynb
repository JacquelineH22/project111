```python
import pandas as pd
import json
```

```python
cells_file = './results/gigatime/cells/cells.csv'
cells = pd.read_csv(cells_file, sep=',', index_col=0, header=0)
cells
```

```python
cells = cells.loc[cells['DAPI']]
cells
```

```python
positive_markers = {
    'T': ['CD3'],
    'CD8+ T': ['CD8', 'CD3'],
    'CD4+ T': ['CD4', 'CD3'],
    'NK': ['CD16'],
    'B': ['CD20'],
    'Plasma': ['CD138'],
    'Monocyte': ['CD14'],
    'Macrophage': ['CD68'],
    'DC': ['CD11c'],
    'Mast': ['Tryptase'],
    'Fibroblasts': ['Transgelin'],
    'Myofibroblast': ['Transgelin', 'Actin-D'],
    'Epithelial': ['CK'],
    'Endothelial': ['CD34'],
}
cell_types = pd.DataFrame(index=cells.index)
for cell_type, markers in positive_markers.items():
    cell_types[cell_type] = cells[markers].all(axis=1)
cell_types
```

```python
cell_types['cell_type'] = None
cell_types.loc[cell_types['T'], 'cell_type'] = 'T'
cell_types.loc[cell_types['CD4+ T'], 'cell_type'] = 'CD4+ T'
cell_types.loc[cell_types['NK'], 'cell_type'] = 'NK'
cell_types.loc[cell_types['B'], 'cell_type'] = 'B'
cell_types.loc[cell_types['Plasma'], 'cell_type'] = 'Plasma'
cell_types.loc[cell_types['Monocyte'], 'cell_type'] = 'Monocyte'
cell_types.loc[cell_types['Macrophage'], 'cell_type'] = 'Macrophage'
cell_types.loc[cell_types['DC'], 'cell_type'] = 'DC'
cell_types.loc[cell_types['Mast'], 'cell_type'] = 'Mast'
cell_types.loc[cell_types['Fibroblasts'], 'cell_type'] = 'Fibroblasts'
cell_types.loc[cell_types['Myofibroblast'], 'cell_type'] = 'Myofibroblast'
cell_types.loc[cell_types['Epithelial'], 'cell_type'] = 'Epithelial'
cell_types.loc[cell_types['Endothelial'], 'cell_type'] = 'Endothelial'
cell_types_unique = cell_types.loc[:, ~cell_types.columns.isin(['CD8+ T', 'CD4+ T', 'Myofibroblast', 'cell_type'])]
cell_types.loc[cell_types_unique.sum(axis=1) != 1, 'cell_type'] = 'Unknown'
cell_types['cell_type'].value_counts()
```

```python
nucleus_seg_json_file = './data/align_he.json'

with open(nucleus_seg_json_file, 'r') as f:
    nucleus_seg_dict = json.load(f)
```

```python
l = []
for i in cell_types.index:
    try:
        cell_types.loc[i, 'x'] = nucleus_seg_dict['nuc'][str(i)]['centroid'][0]
        cell_types.loc[i, 'y'] = nucleus_seg_dict['nuc'][str(i)]['centroid'][1]
    except:
        l.append(i)
cell_types
```

```python
cell_types.to_csv('./results/gigatime/cell_types/cell_types.csv')
cell_types[['x', 'y', 'cell_type']].to_csv('./results/gigatime/results.csv')
```
